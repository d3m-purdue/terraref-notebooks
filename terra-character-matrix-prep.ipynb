### Aggrigation of character matrices for use by Phylogenetic Comparative Methods

The ASR algorithm, phylosignal, and other algorithms need a single row per tree tip.  Since our measurements are per plant per time, they need to be aggregated by species name.  We are averaging the trait values, which works for everything except height, which has been measured at different times throughout the growing season.

```python
import pandas as pd
import numpy as np
```

```python
s4_df = pd.read_csv('/Users/curtislisle/proj/multinet/data/terra_sorghum_s4_leaf_measures.csv')
```

```python
grouped = s4_df.groupby('species')
```

```python
print(len(grouped.agg(np.mean)))
grouped.agg(np.mean).head()
```

```python
grouped.agg(np.mean).to_csv("/Users/curtislisle/proj/multinet/data/terra_sorghum_s4_leaf_means.csv")
```
