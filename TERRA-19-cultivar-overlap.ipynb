### Cultivar Overlap Analysis 

Beward: this was quickly done using the cultivar names.  It is possible the cultivar_id is used consistently across seasons but the names are not.  We found consistent cultivar naming in season 4 and 6, but only a single overlap in season 9, of a cultivar used as a border/control during  season 4 and 6.  

```python
s4df = pd.read_csv('s4_final_heights_and_models.csv')
```

```python
s4df.columns
```

```python
s6df = pd.read_csv('s6_height_and_models_day2.csv')
```

```python
s4df.columns
```

```python
joined = s4df.merge(s6df,how='inner',on='cultivar')
```

```python
# we will use this function later in the notebook
def returnUniqueCounts(dframe):
    return pd.DataFrame.from_records([(col, dframe[col].nunique()) for col in dframe.columns],
                          columns=['Column_Name', 'Num_Unique']).sort_values(by=['Num_Unique'])
```

```python
returnUniqueCounts(joined)
```

```python
s7df = pd.read_csv('s7_height_biomass.csv')
```

```python
joined6_7 = s6df.merge(s7df,how='inner',on='cultivar')
joined6_7.shape
```

```python
returnUniqueCounts(joined6_7)
```
